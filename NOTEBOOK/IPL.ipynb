```python
# —————— ID Columns [viewer_id, session_id] ——————

# viewer_id:
# → Each viewer has a unique ID
# → Used to identify individual users
# → Example: VWR-CD4F36AAD5, VWR-CD4F36AAD5...
# → Total Unique Viewers: 1,999,998

# session_id:
# → Each viewing session has a unique ID
# → One viewer can have multiple sessions
# → Example: SES-4AF6BB1101B2, SES-8AA513B145D4...
# → Used to track individual watch sessions

print("—————— ID Columns Analysis ——————")
print(f"Total Unique Viewers  : {df['viewer_id'].nunique():,}")
print(f"Total Unique Sessions : {df['session_id'].nunique():,}")
print(f"\nSample viewer_id  : {df['viewer_id'].head(3).values}")
print(f"Sample session_id : {df['session_id'].head(3).values}")
```

```python
# —————— Numerical Columns Analysis ——————

# watch_duration_mins:
# → How many minutes a viewer watched the match
# → Range: 0 to 600 mins (full match = ~600 mins)
# → Used to measure viewer engagement

# ads_watched:
# → How many ads a viewer watched completely
# → Range: 0 to 50
# → Higher value = more engaged viewer

# ads_skipped:
# → How many ads a viewer skipped
# → Range: 0 to 50
# → Higher value = less interested in ads

# buffering_events:
# → How many times buffering occurred during watching
# → Range: 0 to 20
# → Higher value = poor streaming experience

# user_rating:
# → Rating given by viewer after watching (1-5)
# → ~40% viewers did not give rating (NaN)
# → Range: 1 to 5

print("—————— Numerical Columns Stats ——————\n")
numerical_cols = ['watch_duration_mins', 'ads_watched', 
                  'ads_skipped', 'buffering_events', 'user_rating']

for col in numerical_cols:
    print("="*50)
    print(col.upper())
    print("="*50)
    print(f"Mean    : {df[col].mean():.2f}")
    print(f"Median  : {df[col].median():.2f}")
    print(f"Std Dev : {df[col].std():.2f}")
    print(f"Min     : {df[col].min():.2f}")
    print(f"Max     : {df[col].max():.2f}")
    print()
```

```python
# 1) WATCH_DURATION_MINS:

# Mean   : 148 mins  → Average viewer 2.5 hours dekha
# Median : 165 mins  → 50% log 165+ mins dekhe
# Min    : 5 mins    → Kuch log sirf 5 min dekh ke chale gaye
# Max    : 240 mins  → Kuch log poora match dekhe (4 hours)

# ✅ Insight: Mean < Median — matlab kuch log bahut kam time dekhe jinhone average neeche khicha

# 2) ADS_WATCHED:

# Mean   : 7 ads  → Average viewer ne 7 ads dekhe
# Median : 7 ads  → Bilkul balanced data
# Min    : 0 ads  → Kuch ne ek bhi nahi dekha
# Max    : 14 ads → Kuch ne sabse zyada dekhe

# ✅ Insight: Mean = Median — perfectly balanced data hai

# 3) ADS_SKIPPED:

# Mean   : 3.5 ads → Average 3-4 ads skip kiye
# Median : 3 ads
# Min    : 0       → Kuch ne ek bhi skip nahi kiya
# Max    : 14      → Kuch ne saare skip kar diye

# ✅ Insight: Log ads_watched (7) se ads_skipped (3.5) ka ratio — 50% ads skip ho rahe hain

# 4) BUFFERING_EVENTS:

# Mean   : 0.60 → Average less than 1 buffering
# Median : 0.00 → 50% logon ko ek bhi buffering nahi aayi!
# Min    : 0    → No buffering
# Max    : 3    → Maximum 3 baar buffering

# ✅ Insight: Median 0 hai — matlab majority logon ka streaming experience bilkul smooth tha!

# 5) USER_RATING:

# Mean   : 4.0 → Average rating 4 out of 5
# Median : 4.0 → Perfectly balanced
# Min    : 3.0 → Kisi ne bhi 3 se kam nahi diya!
# Max    : 5.0 → Kuch ne 5 star diya

# ✅ Insight: Min 3 hai — matlab jo bhi rating diya unhone positive hi diya — content achha tha!
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 7))
fig.suptitle('Match Related Columns Distribution',
             fontsize=16, fontweight='bold')

# Chart 1 - Match Winner
match_winner_counts = unique_match_winner['match_winner'].value_counts()
axes[0].bar(match_winner_counts.index,
            match_winner_counts.values,
            color='orange')
axes[0].set_title('Match Winner', fontweight='bold')
axes[0].set_xlabel('Team')
axes[0].set_ylabel('Wins')
axes[0].tick_params(axis='x', rotation=45)

# Chart 2 - Venue wise Matches
unique_venue_counts = df[['match_no','venue']].drop_duplicates()
venue_counts=unique_venue_counts["venue"].value_counts()
axes[1].bar(venue_counts.index,
            venue_counts.values,
            color='blue')
axes[1].set_title('Venue Wise Matches', fontweight='bold')
axes[1].set_xlabel('Venue')
axes[1].set_ylabel('Matches')
axes[1].tick_params(axis='x', rotation=90)

# Chart 3 - Team wise Matches Played
unique_team1_counts = df[['match_no','team1']].drop_duplicates()
team1_counts = unique_team1_counts["team1"].value_counts()
unique_team2_counts = df[['match_no','team2']].drop_duplicates()
team2_counts = unique_team2_counts["team2"].value_counts()
total_matches = (team1_counts.add(team2_counts, fill_value=0)).sort_values(ascending=False)
axes[2].bar(total_matches.index,
            total_matches.values,
            color='green')
axes[2].set_title('Team Wise Matches Played', fontweight='bold')
axes[2].set_xlabel('Team')
axes[2].set_ylabel('Matches Played')
axes[2].tick_params(axis='x', rotation=45)

team1_counts.add(team2_counts, fill_value=0)

plt.tight_layout()
plt.show()
```
